```python
# Basic GDP scripting calculator
C = int(input("Consumption:"))
I = int(input("Investment:"))
G = int(input("Government Expenditure:"))
X = int(input("Export:"))
M = int(input("Imports:"))

gdp = C + I + G + (X - M)
print("Growth Domestic Product:", gdp)
```

**Another Way**

```python
# Prompt the user to enter values for C, I, G, X, M
input_values = input("Enter the values for C, I, G, X, M (separated by space): ")

# Split the input string into a list of values
values = input_values.split()

# Check if there are five numeric values
if len(values) != 5 or not all(value.isdigit() for value in values):
    print("Error: Please enter valid numeric values for C, I, G, X, and M.")
    exit()

# Convert the values to integers
C, I, G, X, M = map(int, values)

# Calculate GDP using the formula: GDP = C + I + G + (X - M)
GDP = C + I + G + (X - M)

# Output the calculated GDP
print("Gross Domestic Product (GDP):", GDP)
```
