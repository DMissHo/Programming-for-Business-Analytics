**GDP Calculator**

```python
# Basic GDP scripting calculator
a = int(input("Consumption:"))
b = int(input("Investment:"))
c = int(input("Government Expenditure:"))
d = int(input("Export:"))
e = int(input("Imports:"))

gdp = a + b + c + d - e
print("Growth Domestic Product:", gdp)
```
