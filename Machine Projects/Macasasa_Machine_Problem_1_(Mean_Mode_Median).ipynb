```python
# Creating Data
data = "1, 10, 15, 20, 25, 30"
```

```python
# Splitting the dataset into lines
lines = data.split('\n')
```

```python
data_list = [] # Creating a list to store the numeric data
for line in lines: # Iterating through each line and splitting by commas, converting to integers
    if line:
        data = list(map(int, line.split(',')))
        data_list.append(data)

print("List of scores:", data_list)
```

```python
# creating scripts
def calculate_mean(data): # scripting code for mean
    return sum(data) / len(data)
means = sum(data) / len(data)


# scripting code for median
def calculate_median(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    if n % 2 == 0:
        middle1 = sorted_data[n // 2 - 1]
        middle2 = sorted_data[n // 2]
        return (middle1 + middle2) / 2
    else:
        return sorted_data[n // 2]
median = calculate_median(data)

# scripting code for mode
def calculate_mode(data):
    count_dict = {}
    for value in data:
        if value in count_dict:
            count_dict[value] += 1
        else:
            count_dict[value] = 1

mode_values = calculate_mode(data)

# Displaying Values
print("Mean values:", means)
print("Median:", median)
print("Mode:", mode_values)
```

# Other simple way


```python
from statistics import mean, median, mode

# Given dataset
data_list = [2, 3, 3, 4, 6]

# Calculating the mean
mean_value = mean(data_list)

# Calculating the median
median_value = median(data_list)

# Calculating the mode
mode_value = mode(data_list)

# Output
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```
